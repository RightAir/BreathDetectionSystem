```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Method
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
axes.plot(x, y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes1.plot(x, y)
```

```python
axes1.plot(y, x)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# axes.plot(x, y)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2))

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = 'X Squared')
ax.plot(x, x**3, label = 'X Cubed')

ax.legend(loc = (0.1, 0.1))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'purple', lw = 2, ls = '--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
